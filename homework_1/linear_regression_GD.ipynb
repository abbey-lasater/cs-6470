```python
import numpy as np
import matplotlib.pyplot as plt
```

Data Prep

```python
# replace data with that given from homework
X = np.array([[1, 1], [1, 2], [2, 2], [2, 3]])

y = np.dot(X, np.array([1, 2])) + 3
```

```python
plt.scatter(X[:, 1], y)
```

Model - Linear Regression Using Gradient Descent

```python
# setting parameters
# slope
m = 0
# y -int
c = 0 

# learning rate
L = 0.01

# number of iterations
n_iterations = 1000

# number of x instances
n_instances = X.shape[0]
```

```python
# Initialize weights and bias
m = np.zeros(X.shape[1])  # one weight per feature
c = 0

for i in range(n_iterations):
    y_pred = np.dot(X, m) + c
    d_m = (-2/n_instances) * np.dot(X.T, (y - y_pred))  # vectorized derivative w.r.t. m
    d_c = (-2/n_instances) * np.sum(y - y_pred)         # derivative w.r.t. c
    m = m - L * d_m
    c = c - L * d_c

print(m, c)
```

```python
# making predications
y_pred = m*X + c
```

```python
# Visualizing predictions using the second feature (X[:, 1])
plt.scatter(X[:, 1], y, label='Actual')

# Sort values for a smooth line
sorted_idx = np.argsort(X[:, 1])
plt.plot(X[sorted_idx, 1], y_pred[sorted_idx], color='red', label='Predicted')

plt.xlabel('Second feature')
plt.ylabel('Target')
plt.legend()
plt.show()
```

notes

Gradient descent is a mathematical technique that iteratively finds the weights and bias that produce the model with the lowest loss. Gradient descent finds the best weight and bias by repeating the following process for a number of user-defined iterations.

The model begins training with randomized weights and biases near zero, and then repeats the following steps:

Calculate the loss with the current weight and bias.
Determine the direction to move the weights and bias that reduce loss.
Move the weight and bias values a small amount in the direction that reduces loss.
Return to step one and repeat the process until the model can't reduce the loss any further.
